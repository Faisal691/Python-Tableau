## Read Transactions Data

```python
import pandas as pd
data=pd.read_csv('transaction.csv',sep=';')
data
```

## Summary of Data

```python
data.info()
```

### Cost Per transaction column calculation 

```python
CostPerItem=data['CostPerItem']
CostPerItem
```

```python
No_of_items_bought=data['NumberOfItemsPurchased']
No_of_items_bought
```

```python
# Cost per transaction = no of items purchased times cost of each item.
CostPerTransaction=CostPerItem*No_of_items_bought
CostPerTransaction
```

### Cost Per Transactions column creation

```python
# Lets make a new column for this ,in our data


data['CostPerTransaction']=CostPerTransaction
data
```

### Sales per Transaction Calculations and Column Creation

```python
SalesPerTransaction=data['SellingPricePerItem']*data['NumberOfItemsPurchased']
SalesPerTransaction
```

```python
# Create a column for Sales per Transaction

data['SalesPerTransaction']=SalesPerTransaction
data.info()
```

### Profit and Profit% calculations

```python
data['ProfitPerTransaction']=data['SalesPerTransaction']-data['CostPerTransaction']
data['ProfitPerTransaction']
```

```python
# Profit % = (sp-cp)/cp

data['Markup']=data['ProfitPerTransaction']/data['CostPerTransaction']
data['Markup']
```

```python
data.info()
```

```python
# Rounding Markup to two decimal places.

RoundMarkup=round(data['Markup'],2)
RoundMarkup

#now update the new Markup column
data['Markup']=RoundMarkup
data['Markup']
```

### Combining Data Fields......

```python
# Here we are combining Day, Month and Yead column to a single Date column.
# We will create a new column of type -> day-month-year.
# But as Day and year is integer , we cannot concatenate '-' in it , hence we need to change the data type as str.
# lets de that.

day=data['Day'].astype(str)
print(day.dtype)

year=data['Year'].astype(str)

Date=day+"-"+data['Month']+"-"+year
Date

# Add Date as a new column in dataset.
data['Date']=Date
data
```

```python
''' To View Specific pars of data , 
we use a special function called iloc.
iloc fives contents of specific rows..

ex. data.iloc[0] -> gives all data from row 1
    data.iloc[1:6] -> gives all data from row 1 to row 5.
    data.iloc[-5] -> gives data of last 5 rows.
    data.head() -> gives first 5 rows.
    data.head(n) -> gives first n rows........
    
    etc..   '''

```

### Splitting ClientKeywords column to 3 different colunms....

```python
# We use split() func. for this.
# *** Ye Format Yaad Rakhna ***

split_col=data['ClientKeywords'].str.split(',' , expand=True)
split_col
```

```python
# Now lets add these columns individually to our dataset......

data['ClientAge']=split_col[0]
data['OrganisationType']=split_col[1]
data['LengthOfContract']=split_col[2]
data.info()
```

```python
data['ClientAge']=data['ClientAge'].str.replace('[','')
data['LengthOfContract']=data['LengthOfContract'].str.replace(']','')
data.head()
```

## Bringing in new dataset.....

```python
df=pd.read_csv('value_inc_seasons.csv',sep=';')
df
```

```python
# Now we will merge the two data fields.
#format : dataFrame=pd.merge(df1,df2,on='key')
# Key is the common field between the two dataframe.

data=pd.merge(data,df,on='Month')
data.head()
```

```python
# Now we don't need some columns, like Day,Month,Year and ClientKeywords 
# Becoz we have already optimized those fields.
# Python has a drop() func. that is used to remove certain fields.

data=data.drop('Day',axis=1)
# when axis=1 , it ,means we need to drop a column.
# axis=0 , means we drop a row.

# ** we can also drop multiple columns at a same time usig [] brabkets. **

data=data.drop(['Month','Year','ClientKeywords'],axis=1)
data.info()
```

```python
## Now our data is cleaned , lets create a new csv file and save it.

data.to_csv('ValueInc_cleaned.csv')
```
